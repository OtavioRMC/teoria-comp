```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from random import randint
```

1. Carregar o Dataset

```python
facebook = pd.read_csv("/content/facebook_combined.txt.gz",
                       compression="gzip",
                       sep=" ",
                       names=["start_node","end_node",]
                       )
```

```python
facebook
```

```python
G = nx.from_pandas_edgelist(facebook, "start_node", "end_node")
```

2. Análisde Descritiva

```python
# Número de nodes no grafo
G.number_of_nodes()
```

```python
# Número de arestas no grafo
G.number_of_edges()
```

```python
# Média do grau de cada nó
np.mean([d for _, d in G.degree()])
```

```python
# Calcula os comprimentos dos caminhos mais curtos entre todos os pares de nós
shortest_path_lengths = dict(nx.all_pairs_shortest_path_length(G))
```

```python
shortest_path_lengths[0][42]
```

O diâmetro de um grafo é a maior distância mínima entre qualquer par de nós no grafo. Em outras palavras, é a distância do caminho mais curto entre os nós mais distantes.

```python
# Calcua o Diametro médio para atravessar de um nó para o outro
# nx.eccentricity = Caminho mais longo para o nó mais distante
diameter = max(nx.eccentricity(G, sp=shortest_path_lengths).values())
diameter
```

```python
# Média dos caminhos mais curtos
average_path_lengths = [
    np.mean(list(spl.values())) for spl in shortest_path_lengths.values()
]
# Média de todos os nós
np.mean(average_path_lengths)
```

Para atravessar entre um nó e outro em média 3.6 arestas são percorridas

Distribuição dos caminhos mais curtos na rede.

```python
# Sabemos o comprimento máximo do caminho mais curto (diametro), criar array que guarda
# os valores de 0 até 8
path_lengths = np.zeros(diameter + 1, dtype=int)

# Frquência do caminho mais curto observada
for pls in shortest_path_lengths.values():
    pl, cnts = np.unique(list(pls.values()), return_counts=True)
    path_lengths[pl] += cnts

# Converte para porcentagem
freq_percent = 100 * path_lengths[1:] / path_lengths[1:].sum()

# Plota a  distribuição de frequência  (ignora caminhos de de 0) como %
fig, ax = plt.subplots(figsize=(15, 8))
ax.bar(np.arange(1, diameter + 1), height=freq_percent)
ax.set_title(
    "Distribuição dos caminhos mais curtos na Rede", fontdict={"size": 35}, loc="center"
)
ax.set_xlabel("Comprimento do Caminho mais curto", fontdict={"size": 22})
ax.set_ylabel("Frequência (%)", fontdict={"size": 22})
```
